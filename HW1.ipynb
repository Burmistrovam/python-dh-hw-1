### Задание 1
Напишите программу, которая по данному числу n от 1 до 9 выводит на экран n пингвинов.

```python
number = int(input('Количество пингвинов от 1 до 9: '))
if number < 1:
    print("Некого рисовать!")
elif number > 9:
    print("Слишком много пингвинов!")
else:
    print(number * '   _~_    ')
    print(number * '  (o o)   ')
    print(number * ' /  V  \\  ')
    print(number * '/(  _  )\\ ')
    print(number * '  ^^ ^^   ')
```

### Задание 3
Возьмите достаточно длинную строку, хотя бы 20 символов.

Используя срезы (конструкции с квадратными скобочками и двоеточиями при необходимости, которые были пройдены на 4 семинаре ([i],[a:b], [a:b:step]; работают со строками также, как со списком)), сделайте следующее:

Сначала выведите третий символ этой строки.
Во второй строке выведите предпоследний символ этой строки.
В третьей строке выведите первые пять символов этой строки.
В четвертой строке выведите всю строку, кроме последних двух символов.
В пятой строке выведите все символы с четными индексами (считая, что индексация начинается с 0, поэтому символы выводятся начиная с первого).
В шестой строке выведите все символы с нечетными индексами, то есть начиная со второго символа строки.
В седьмой строке выведите все символы в обратном порядке.
В восьмой строке выведите все символы строки через один в обратном порядке, начиная с последнего.
В девятой строке выведите длину данной строки.

```python
sentence3 = input('Введите предложение: ')
print(sentence3[2])     # третий символ этой строки
print(sentence3[-2])    # предпоследний символ этой строки
print(sentence3[:5])    # первые пять символов этой строки
print(sentence3[:-2])   # всю строку, кроме последних двух символов.
print(sentence3[0::2])  # все символы с четными индексами (считая, что индексация начинается с 0, поэтому символы выводятся начиная с первого).
print(sentence3[1::2])  # все символы с нечетными индексами, то есть начиная со второго символа строки.
print(sentence3[::-1])  # все символы в обратном порядке.
print(sentence3[-1::-2])  # все символы строки через один в обратном порядке, начиная с последнего.
print(len(sentence3))
```

### Задание 4
Вася хочет проверить, верно ли, что данная текстовая строка читается слева направо и справа налево одинаково. Помогите Васе написать такую программу. Возьмите два слова из строчных букв. Для первого слова, которое должно читаться одинаково, выведите YES, а для второго слова, которое должно читаться неодинаково, -- NO.

```python
word1 = input('Введите слово №1: ')
if word1[:] == word1[::-1]:
    print('YES')
else:
    print('NO')
    
word2 = input('Введите слово №2: ')
if word2[:] == word2[::-1]:
    print('YES')
else:
    print('NO')
```
